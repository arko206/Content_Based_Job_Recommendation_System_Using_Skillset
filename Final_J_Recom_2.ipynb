## Stemming Process is Done

```python
def stem(text):
    y = []
    
    for i in text.split(): #convert each string in text to list
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
indices = pd.Series(new_df.index, index=new_df['skills']).drop_duplicates()
```

## Recommendation Function

```python
def give_rec(skill_set, similarity = similarity):
  #Get the index corresponding to original title
  idx = indices[skill_set]

  #Get pairwise Similarity Scores
  sim_scores = list(enumerate(similarity[idx]))

  #Sort the Scores in Descending Order
  sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)

  #Scores of the 10 most similar Skills
  sim_scores = sim_scores[1:11]

  #Job Title indices
  skills_indices = [i[0] for i in sim_scores]
  #df1 = pd.DataFrame(data = jobs_cleaned_data[skills_indices], columns=['Title','Link','Name'])
  data1 = jobs_cleaned_data['jobtitle'].iloc[skills_indices]
  data2 = jobs_cleaned_data['company'].iloc[skills_indices]
  data3 = jobs_cleaned_data['advertiserurl'].iloc[skills_indices]

  #Top 10 most similar Jobs
  return data1,data2,data3
```
